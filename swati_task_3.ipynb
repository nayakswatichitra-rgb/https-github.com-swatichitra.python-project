```python
class Greeting:
    @staticmethod
    def say_hello():
        print("Hello!")

Greeting.say_hello()
```

```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

result = Calculator.add(10, 20)
print(result)
```

```python
class Number:
    @staticmethod
    def is_even(num):
        return num % 2 == 0

print(Number.is_even(8))
print(Number.is_even(7))
```

```python
class Temperature:
    @staticmethod
    def celsius_to_fahrenheit(c):
        return (c * 9 / 5) + 32

print(Temperature.celsius_to_fahrenheit(25))
```

```python
class Validator:
    @staticmethod
    def is_valid_email(email):
        return "@" in email and "." in email

print(Validator.is_valid_email("abc@gmail.com"))
print(Validator.is_valid_email("abcgmail"))
```

```python
class Geometry:
    @staticmethod
    def circle_area(radius):
        pi = 3.14159
        return pi * radius * radius

print(Geometry.circle_area(5))
```

```python
class Student:
    school = "ABC School"

    @classmethod
    def get_school(cls):
        return cls.school

print(Student.get_school())
```

```python
class Employee:
    company = "TechCorp"

    @classmethod
    def change_company(cls, new_company):
        cls.company = new_company

print(Employee.company)

Employee.change_company("OpenAI")

print(Employee.company)
```

```python
class Car:
    count = 0

    def __init__(self):
        Car.count += 1

    @classmethod
    def total_cars(cls):
        return cls.count

Car()
Car()
Car()

print(Car.total_cars())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def from_string(cls, data):
        name, age = data.split("-")
        return cls(name, int(age))

p = Person.from_string("Alice-25")

print(p.name)
print(p.age)
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

r = Rectangle(5, 4)
print(r.area())
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Balance:", self.balance)

acc = BankAccount(1000)
acc.deposit(500)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def can_vote(self):
        if self.age >= 18:
            print(self.name, "can vote")
        else:
            print(self.name, "cannot vote")

p = Person("John", 17)
p.can_vote()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def increase_salary(self, amount):
        self.salary += amount
        print("New Salary:", self.salary)

e = Employee("David", 30000)
e.increase_salary(5000)
```

```python
class Car:
    def __init__(self):
        self.speed = 0

    def accelerate(self, value):
        self.speed += value
        print("Current Speed:", self.speed)

car = Car()
car.accelerate(40)
car.accelerate(20)
```
